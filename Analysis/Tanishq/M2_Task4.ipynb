```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data_frame = pd.read_csv("/Users/amitaaneja/Desktop/Data301/course-project-group_6003/data/raw/Gamesales.csv")
data_frame.head(7)  
```

```python
new_data_frame = data_frame.rename(columns = {"Platform": "Video Game Console"})
new_data_frame = data_frame.rename(columns = {"Genre": "Category"})
new_data_frame.dropna(how = 'any')
new_data_frame.drop_duplicates()  

# This code helps drop the NA values and removes the duplicates if any.
```

```python
sns.countplot(y=new_data_frame['Year_of_Release'])
plt.title("Amount of new releases per year")

## helps us understand the year of release
```

```python
print(data_frame.columns)
data_frame['Year_of_Release'].plot(kind='hist', bins=20, figsize=(12,6), facecolor='teal',edgecolor='grey')
##This Histogram helps us answer the most amount of sales that were made in the gap of years 1980 to 2020.
##It happens to be in 2009-2010 period. Rest I made the bar teal in colour because it’s my favourite and had a prominent effect.
```

```python
games = data_frame[['Name', 'Year_of_Release', 'Genre', 'JP_Sales', 'Global_Sales']]
RolePlaying_games = games[games['Genre'] == 'Role-Playing']
print(f'Total Role-Playing game sales in JP: {RolePlaying_games["JP_Sales"].sum()}')
act_game = games[games['Genre'] == 'Action']
print(f'Total Action game sales in JP: {act_game["JP_Sales"].sum()}')
spor_games = games[games['Genre'] == 'Sports']
print(f'Total Sports game sales in JP: {spor_games["JP_Sales"].sum()}')
Platf_games = games[games['Genre'] == 'Platform']
print(f'Total Platform game sales in JP: {Platf_games["JP_Sales"].sum()}')
Race_games = games[games['Genre'] == 'Racing']
Puzz_games = games[games['Genre'] == 'Puzzle']
print(f'Total Puzzle game sales in JP: {Puzz_games["JP_Sales"].sum()}')
print(f'Total Racing game sales in JP: {Race_games["JP_Sales"].sum()}')
Adven_games = games[games['Genre'] == 'Adventure']
print(f'Total Adventure game sales in JP: {Adven_games["JP_Sales"].sum()}')


## Helps calculate average sale per genre in japan
```

```python
# Comparing the sales of different genres of video games in Japan

sns.violinplot(x = 'Genre', y = 'JP_Sales', data = data_frame)
plt.xticks(rotation = 90)
plt.title("Sales of different genres in Japan")

## This creates a violin plot assists us with the most popular genre in Japan, which was Role playing games 
## followed by platform games
```
